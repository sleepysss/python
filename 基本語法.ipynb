```python
#set

'''

無序的集合,不會包含重複的資料

'''

########
# 建立 #
#######

s1 = set() # set={}會建立空的dict,而不是set
s2 = set((2, 3, 1)) # 帶入 tuple 資料來初始化
s3 = {1, 2, 3, 4, 5}

#set comprehension
titles = "Learn Code With Mike"
result = {letter for letter in titles if letter == "e"}

############
# 新增刪除 #
##########
s1.add(6)
s1.remove(5)


########
# 尋找 #
#######

fruits = {'apple','banana','orange','lemon'}
result = 'apple' in fruits


########
# 交集 #
#######

fruits1 = {'apple','banana','orange','lemon'}
fruits2 = {'tomato','apple','banana'}
print(fruits1 & fruits2)

nums1 = {1,2,3,4,5}
nums2 = {2,4,6,8,10}
print(nums1.intersection(nums2))


########
# 聯集 #
#######

fruits1 = {'apple','banana','orange','lemon'}
fruits2 = {'tomato','apple','banana'}
print(fruits1 | fruits2)

nums1 = {1,2,3,4,5}
nums2 = {2,4,6,8,10}
print(nums1.union(nums2))


########
# 差集 #
#######

fruits1 = {'apple','banana','orange','lemon'}
fruits2 = {'orange','lemon','tomato'}
print(fruits1 - fruits2) #得到fruits1 set 元素且不包含 fruits2 set 的元素

nums1 = {1,2,3,4,5}
nums2 = {4,5,6,7,8}
print(nums1.difference(nums2))


############
# 對稱差集 #
##########

#兩個集合的對稱差是只屬於其中一個集合，而不屬於另一個集合的元素組成的集合。
#{1,2,3},{3,4}的對稱差極為{1,2,4}

fruits1 = {'apple','banana','orange','lemon'}
fruits2 = {'orange','lemon','tomato'}
print(fruits1 ^ fruits2) #{'tomato', 'banana', 'apple'}

nums1 = {1,2,3,4,5}
nums2 = {4,5,6,7,8}
print(nums1.symmetric_difference(nums2))
```
